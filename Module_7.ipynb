```python
# seek function
# by using seek we can move index both left and right

with open("./sample_data/sample.txt","r") as file:
  print(file.tell())
  print(file.read(5))
  print(file.tell())
  file.seek(0)
  print(file.tell())
  print(file.read())
```

```python
# from os import linesep
# Practice
# get the text from sample.txt and count the following
# numbers of lines
# numbers of words
# numbers of characters
# and save it to the counter_of string.txt file
from functools import reduce
with open("./sample_data/sample.txt","r") as file:
  lines=file.readlines()
  total_lines=len(lines)

  word=list(map(lambda x:len(x.split()),lines))
  # print(word)
  total_num=reduce(lambda x,y:x+y,word)
  # print(total_num)

  # cleaning process
  # new line delete
  lines=list(map(str.strip,lines))
  # space delete
  lines=list(map(lambda x:x.replace(" ",""),lines))
  # character how much
  num_of_char=list(map(lambda x:len(x),lines))
  total_num_of_char=reduce(lambda x,y:x+y,num_of_char)

with open("./sample_data/counter_of_string.txt","w")as file:
  file.write(f"total line: {total_lines}\ntotal number of words:{total_num}\ntotal character:{total_num_of_char}")





```
