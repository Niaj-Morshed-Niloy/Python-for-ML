```python
set={2,3,5,7,8,4,2}
sum=0
for element in set:
    sum+=element
print(sum)
```

```python
# method in set

s={6,1,2,3,4,3,5}
s.add(10)
s.pop()
s.remove(6)

print(s)
```

```python
a={1,2,3}
b={2,3,4}
s=a.union(b)
k=a.intersection(b)
print(s)
print(k)
```

```python
# joint isdisjoint
a={1,2,3}
b={1,2,3,5,6,4}
print(a.isdisjoint(b))
print(a.issubset(b))
```

```python
# view method in dictionary

dic={"name":"phitron",
     "age":20,
     "address":"Dhaka",
     "numbers":[10,20,30]
    }

keys=dic.keys()
print(keys)
dic["math_marks"]=30
print(keys)

values=dic.values()
print(values)

dic["math_marks"]=40
print(values)

items=dic.items()
print(items)
```

```python
# iteration
for key,value in dic.items():
    print(key,value)
```

```python
# Codeforces j.count letter proble,
s=input().strip()
# s=sorted(input().strip()) #thats also valid
count={}
for word in s:
    count[word]=count.get(word,0)+1
for k,v in sorted(count.items()):
    print(f"{k}:{v}")
```

```python
# or
s=input()
st=sorted(s.strip())
count={}
for word in st:
    count[word]=count.get(word,0)+1

for k,v in count.items():
    print(f"{k}:{v}")
```
