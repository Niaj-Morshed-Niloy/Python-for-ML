```python
def square_addition(a,b):
    a=a**2
    b=b**2
    return a+b
ans=square_addition(2,3)+10
print(ans)
```

```python
def square_addition(*args):
    summation=0
    for i in args:
        i=i**2
        summation+=i
    return summation
ans=square_addition(10,20,30,40,50)
print(ans)
```

```python
# Max Min problem in codeforces using lambda

from functools import reduce
n=int(input())
a=list(map(int, input().split()))
maxii=reduce(lambda x,y:x if x>y else y,a)
minii=reduce(lambda x,y:y if y<x else x,a)
print(minii,maxii)
```

```python
# mean problem in codeforces using reduce
from functools import reduce
inp=int(input())
lara=list(map(float, input().split()))
avg=reduce(lambda x,y:x+y,lara)
mean=avg/inp
print(f"{mean:.6f}")
```
