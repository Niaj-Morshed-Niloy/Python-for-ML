```python
#Calculate the total number spent by each user(using loop:)
user_data=[
    {"user_id":1,"name":"Niaj","orders":[{"orders_id":"A1","amount":200},{"orders_id":"A2","amount":300},{"orders_id":"A3","amount":500}]},
    {"user_id":2,"name":"Anon","orders":[{"orders_id":"B1","amount":5},{"orders_id":"B1","amount":10},{"amount":70}]},
    {"user_id":3,"name":"charlie","orders":[]}

]
for user in user_data:
  total_spent=0
  for order in user.get("orders",[]):
    total_spent+=order.get("amount",0)
  print(f"{user.get('name','N/A')}: ${total_spent}")

  #get a list of all order
  all_order_amounts=[order.get("amount",0)for user in user_data for order in user.get("orders",[])]
  print(all_order_amounts)

  #dictionary mapping user_id to the number of orders

  orders_count_by_user={user.get("user_id"):len(user.get("orders",[])) for user in user_data}
  print(orders_count_by_user)

```
