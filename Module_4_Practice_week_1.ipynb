```python
inp=input()
number=inp.split()
x=float(number[0])
min_v=float(number[1])
max_v=float(number[2])
norm = (x - min_v) / (max_v - min_v)
print(f"{norm:.2f}")
```

```python
N=int(input())
target=float(input())
total=0
for i in range(N):
  loss_i=float(input())
  total=total+loss_i
average_loss=total/N
if(average_loss<=target):
  print("PASS")
else:
  print("RETRY")
```

```python
N=int(input())
YES_count=0
NO_count=0
for i in range (N):
  vote=input()
  if vote=="YES":
    YES_count+=1
  else:
    NO_count+=1
if YES_count >= NO_count:
  print("ACCEPT")
else:
  print("REJECT")


```

```python
users_message=input()
if "happy" in users_message or "joy" in users_message or "smile" in users_message:
  print("Happy Mood")
elif "sad" in users_message or "cry" in users_message or "angry" in users_message:
  print("Sad Mood")
else:
  print("Neutral Mood")
```

```python
pixels = list(map(int, input().split()))
avg = sum(pixels) / len(pixels)
if avg < 85:
    print("Dark Image")
elif 85<=avg<=170:
    print("Normal Image")
else:
    print("Bright Image")
```

```python
predict=input().split()
total=len(predict)

countA=predict.count('A')
countB=predict.count('B')

avg_countA=countA/total
avg_countB=countB/total

if (avg_countA>0.7) or (avg_countB>0.7):
  print("Biased Model")
else:
  print("Fair Model")

```

```python
user_message=input()
keyword=["ai", "data", "model", "learn", "train", "neura"]
count=sum(word in user_message for word in keyword)
if count>=2:
  print("AI Detected")
else:
  print("Not AI Related")
```
