```python
# Codeforce: K max and min problem
n=input()
inp=n.split()
x=int(inp[0])
y=int(inp[1])
z=int(inp[2])
min=x
max=x
if min>y:
    min=y
if min>z:
    min=z

if max<y:
    max=y
if max<z:
    max=z

print(min,max)
```

```python
#c. even odd positive negative

n=int(input())
inp=input()
number=inp.split()
positive=0;negative=0;even=0;odd=0
for i in range (n):
    x=int(number[i])

    if x>0:
        positive+=1
    else:
        negative+=1
    if x%2==0:
        even+=1
    else:
        odd+=1

print("Positive",positive)
print("Negative",negative)
print("Odd",odd)
print("Even",even)




```

```python
pixels=list(map(int, input().split()))
avg=sum(pixels) / len(pixels)
if avg < 85:
    print("Dark Image")
elif 85<=avg<=170:
    print("Normal Image")
else:
    print("Bright Image")
```
