```python
# Create a 1D, 2D, and 3D NumPy array and print their dimensions and shapes.
import numpy as np
arr1=np.array([10,20,30])
arr2=np.array([[10,20,30],
               [40,50,60]
              ])
arr3=np.array([[[10,20,30],
               [40,50,60]],

               [[10,20,30],
               [40,50,60]],

               [[10,20,30],
               [40,50,60]]])
print(arr1)
print(arr2)
print(arr3)
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)
```

```python
""" Create:
an array of zeros of shape (3,3)
an array of ones of shape (2,5)
an array filled with 7"""

arr=np.zeros((3,3))
print(arr)

arr_1=np.ones((2,5))
print(arr_1)

arr_2=np.full_like((arr1),7)
print(arr_2)
arr_2=np.full_like((arr2),7)
print(arr_2)
arr_2=np.full_like((arr3),7)
print(arr_2)
```

```python
# Create an array of even numbers between 10 and 50 using arange.
arr=np.arange(2,51,2).reshape(5,5)
print(len(arr))
print(arr)
```

```python
"""Create a 4×4 matrix from 0–15. Then , extract:
the first row
the last column
a 2×2 subarray from the center"""

arr=np.arange(0,16,1).reshape(4,4)
print(arr)

row_0=arr[0:1,]
print(row_0)

col_last=arr[::,3:]
print(col_last)

portion=arr[1:3,1:3]
print(portion)
print(portion.shape)
```
